```python
from hdna import * 
import nupack as nu
import pandas as pd 
```

```python
# Import experimental data from Hertel 
expdata = pd.read_csv('../data/herteldata.csv', names=['seq', 'expvalue'])
# Clean the dataframe 
expdata = expdata.drop(0)
expdata['expvalue'] = ['{:e}'.format(float(e)) for e in expdata['expvalue']]

limit = len(expdata)
torun = expdata.copy().iloc[:limit]
torun['index'] = torun.index 
torun.set_index(torun['seq'], inplace=True)
```

```python
seq = torun.iloc[0]['seq']
```

```python
model = Model('dna', '3D')
A = Strand(model, seq)
B = A.complementary()
G = Geometry(120, 270)
K = Kinetwork(model, A, B, G)
S = Simulator(model, K)
```

```python
K.save_graph('checht')
```

```python
run = S.ensemble()
```

```python
fpts = S.fpts(run)
```

```python
import plotly.graph_objects as go
from scipy.stats import expon
a, b = expon.fit(fpts)
import numpy as np 
trace1 = go.Histogram(x=fpts, nbinsx=60, histnorm='probability density')

x = np.linspace(0, 1e-5, 100)
trace2 = go.Scatter(x=x, y=expon.pdf(x, loc=a, scale=b))
dati = [trace1, trace2]
fig = go.Figure(data = dati)
fig.update_xaxes(exponentformat="e")
fig.update_yaxes(exponentformat="e")
print(a, b)
print(S.mfpt(run))
fig.show()
```
