```python
import numpy as np 
import pandas as pd 
from hdna import *
```

```python
model = Model('dna', '2D')
# A = Strand.random(model, 10)
# B = A.complementary()
A = Strand(model, 'TAATCGAAAG')
B = A.complementary()
kinetwork = Kinetwork(model, A, B, 3)
opts = Options()
geo = Geometry(120, 120)
kinetics = Kinetics(model, kinetwork, geo)

```

```python
print(A.sequence+'\n'+B.sequence)
```

```python
options = Options(method='direct', runtime=1e-5, Nmonte=500)
simulatore = Simulator(model, kinetwork, kinetics, options=options)
E = simulatore.ensemble()
E = simulatore.ensemble_to_dataframe(E)
```

```python
tau = simulatore.mfpts(E, converted=True)
print(tau)
```
