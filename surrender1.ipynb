```python
from hdna import *
```

```python
model = Model('dna', '2D')
A = Strand.random(model, 30)
B = A.complementary()
kinetwork = Kinetwork(model, A, B, 3)
```

```python
opts = Options()
geo = Geometry(120, 120)
kinetics = Kinetics(model, kinetwork, geo)
options = Options(method='direct', runtime=1e-6, Nmonte=50000)
simulator = Simulator(model, kinetwork, kinetics, options=options)
```

```python
E = simulator.ensemble()
```

```python
len(E)
```

```python
BS = simulator.biosim

""" Inspect the biosim object """

print(len(BS.species_list) == len(simulator.Graph.nodes())) 
print(int(len(BS.reaction_list)/2) == len(simulator.Graph.edges()))

""" Graph properties match between networkx and BioSimulator """

```
